```python
transactions= []
for i in range(0, len(df)):
  transactions.append( [str(df.values[i,j]) for j in range(0, 20) if str(df.values[i,j]) != '0'] )
```

```python
from apyori import apriori
```

```python
rules =  apriori(transactions = transactions , min_support = 0.003, min_confidence = 0.2, min_lift = 3 ,
                 min_length = 2, max_length = 2)
```

```python
result = list(rules)
```

```python
import pandas as pd
def apriori_to_dataframe(results):
    rows = []
    for result in results:
        for rule in result.ordered_statistics:
            rows.append({
                "Antecedent": ", ".join(rule.items_base),
                "Consequent": ", ".join(rule.items_add),
                "Support": result.support,
                "Confidence": rule.confidence,
                "Lift": rule.lift
            })
    return pd.DataFrame(rows)
```

```python
rules_df = apriori_to_dataframe(result)
rules_df.head()
```
